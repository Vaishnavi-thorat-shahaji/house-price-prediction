```python
col = ['Suburb','Rooms','Type','Price','Method','SellerG','Date','Distance','Postcode','Bedroom2','Bathroom','Car','Landsize','BuildingArea','YearBuilt','CouncilArea','Lattitude','Longtitude','Regionname','Propertycount']

def encode(c):
    le.fit(data[c])
    data[c]= le.fit_transform(data[c])
for c in col:
    encode(c)
```
